$H=AeσT^4$

```python
import sympy as sym

# Definir las variables
A = sym.Symbol('A')
e = sym.Symbol('e')
T = sym.Symbol('T')
σ = 5.67 * 10**-8

# Definir la función H
H = A * e * σ * T**4

# Valores y errores
vA = 0.15
ve = 0.9
vT = 650
eA = 0.01
eE = 0.01
eT = 20

eH= sym.diff(H, A).subs({A: vA, e: ve, T: vT})*eA + sym.diff(H, e).subs({A: vA, e: ve, T: vT})*eE + sym.diff(H, T).subs({A: vA, e: ve, T: vT})*eT
H=H.subs({A: vA, e: ve, T: vT})
print("Funcion H evaluada:",H)
print("Error de H:",eH)
print("Limite Superior:",H+eH)
print("Limite Inferior:",H-eH)

```

```python
import numpy as np
import sympy as sym
```

```python
#Numero base
x0= 1
#Numero objetivo
x1= 3
#Diferencia
h=x1-x0
#Definición de función
x=sym.Symbol('x')
f=25*x**3-6*x**2+7*x+88
#Orden esperado
n=3

aproxTaylor=0

for i in range(n+1):
  aproxTaylor+=sym.diff(f,x,i).subs(x,x0)/sym.factorial(i)*(h)**(i)
  print("Orden",i,":",aproxTaylor)

print("Valor buscado:",f.subs(x,x1))
print("Valor aproximado:",aproxTaylor)

```

##EJERCICIO 2
Serie de Taylor para $ln(x)$

$\ln(x) = (x - 1) - \frac{(x - 1)^2}{2} + \frac{(x - 1)^3}{3} - \frac{(x - 1)^4}{4} + \dots
$

$\sum_{n=1}^{\infty} (-1)^{n+1}⋅\frac{(x - 1)^n}{n}
$

```python
#Numero base
x0= 1
#Numero objetivo
x1= 2.5
#Diferencia
h=x1-x0
#Definición de función
x=sym.Symbol('x')
f=sym.ln(x)
#Orden esperado
n=5

aproxTaylor=0
for i in range(1,n+1):
  rn= (h**i/i)*(-1)**(i+1)
  aproxTaylor+=rn
  print("Orden",i-1,":",aproxTaylor)

print("Valor buscado:",f.subs(x,x1))
print("Valor aproximado:",aproxTaylor)
```

## EJERCICIO 3
$Q = \frac{1}{n} \frac{(B \cdot H)^{\frac{5}{3}}}{(B + 2H)^{\frac{2}{3}}} \sqrt{S}$


```python
import sympy as sym

# Definir las variables
B = sym.Symbol('B')
H = sym.Symbol('H')
S = sym.Symbol('S')
n = sym.Symbol('n')

# Definir la función Q
Q = (1/n)*((B*H)**(5/3)/(B+2*H)**(2/3))*S**(1/2)

# Valores y errores
vB = 20
vH = 0.3
vS = 0.0003
vN = 0.03
eB = 0
eH = 0
eS = vS*0.1
eN = vN*0.1

eQ= abs(sym.diff(Q, S).subs({B: vB, H: vH, S: vS, n: vN}))*eS+abs(sym.diff(Q, n).subs({B: vB, H: vH, S: vS, n: vN}))*eN
Q=Q.subs({B: vB, H: vH, S: vS, n: vN})
print("Funcion Q evaluada:",Q)
print("Error de Q:",eQ)
print("Limite Superior:",Q+eQ)
print("Limite Inferior:",Q-eQ)
```
